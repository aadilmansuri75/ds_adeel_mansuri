```python
def find_col(df, candidates):
  for c in candidates:
    if c in df.columns:
      return c

  cols_lower = {col.lower(): col for col in df.columns}
  for c in candidates:
    if c.lower() in cols_lower:
      return cols_lower[c.lower()]
  return None


hist_path = csv_output_dir / "historical_data.csv"
fg_path = csv_output_dir / "fear_greed_index.csv"

hist = pd.read_csv(hist_path)
fg = pd.read_csv(fg_path)

print("fear & greed dataset head")
display(fg.head(5))

print("historical dataset head")
display(hist.head(5))
```

```python
date_col_fg = find_col(fg, ["Date", "date", "Time", "time"])
class_col_fg = find_col(fg, ["Classification", "classification", "Value", "Sentiment", "sentiment"])
if date_col_fg is None:
  raise ValueError("Could not find date column  in fear & greed dataset. Columns: " + str(list(fg.columns)))

if class_col_fg is None:
  raise ValueError("Could not find classification column in fear greed dataset. Columns: " + str(list(fg.columns)))

fg = fg.rename(columns={date_col_fg: "date", class_col_fg: "classification"})
fg['date'] = pd.to_datetime(fg['date'], errors='coerce').dt.date

fg['classification'] = fg['classification'].astype(str).str.strip().str.title()
fg['classification'] = fg['classification'].replace({'Greed':'Greed', 'Fear':'Fear'})
fg_grouped = fg.groupby('date', as_index=False).agg({'classification': lambda x: x.mode().iat[0] if len(x)>0 else np.nan})

print("\n Unique classification found:", fg_grouped['classification'].unique())
```

```python
col_time = find_col(hist, ["Timestamp IST", "time", "Time", "timestamp", "date"])
col_account = find_col(hist, ["Account", "account", "acct"])
col_symbol = find_col(hist, ["Coin", "symbol", "market"])
col_exec_price = find_col(hist, ["Execution Price", "execution price", "price", "exec_price"])
col_size = find_col(hist, ["Size Tokens", "Size USD", "size", "qty", "quantity"])
col_side = find_col(hist, ["Side", "side", "direction"])
col_start_pos = find_col(hist, ["Start Position", "start_position", "startPos", "startpos"])
col_event = find_col(hist, ["Direction", "event", "Event"])
col_closedPnL = find_col(hist, ["Closed PnL", "closedPnL", "closed_pnl", "pnl", "PnL"])
col_leverage = find_col(hist, ["Leverage", "leverage", "Lev", "lev"])

print("\nDetected columns in historical data:")
print("Time:", col_time)
print("Account:", col_account)
print("Symbol:", col_symbol)
print("Execution Price:", col_exec_price)
print("Size:", col_size)
print("Side:", col_side)
print("Start Position:", col_start_pos)
print("Event:", col_event)
print("Closed PnL:", col_closedPnL)
print("Leverage:", col_leverage)
```

```python
if col_time:
  hist['time_parsed'] = pd.to_datetime(
    hist['Timestamp IST'],
    format='%d-%m-%Y %H:%M',
    errors='coerce'
)
  hist['date'] = hist['time_parsed'].dt.date
  print(hist[['Timestamp IST', 'time_parsed', 'date']].head(10))

  if hist['time_parsed'].isna().all():
    try:
      hist['time_parsed'] = pd.to_datetime(hist[col_time].astype(float), unit='s', errors='coerce')
    except Exception:
      pass
else:
  hist['time_parsed'] = pd.NaT
hist['date'] = hist['time_parsed'].dt.date

if col_size:
  hist['size_num'] = pd.to_numeric(hist[col_size], errors='coerce').fillna(0)
else:
  hist['size_num'] = 0

if col_exec_price:
  hist['price_num'] = pd.to_numeric(hist[col_exec_price], errors='coerce')
else:
  hist['price_num'] = np.nan

if col_closedPnL:
  hist['closedPnL_num'] = pd.to_numeric(hist[col_closedPnL], errors='coerce').fillna(0)
else:
  hist['closedPnL_num'] = 0

if col_leverage:
  hist['leverage_num'] = pd.to_numeric(hist[col_leverage], errors='coerce').fillna(1)
else:
  hist['leverage_num'] = 1

hist['is_win'] = hist['closedPnL_num'] > 0
hist['abs_pnl'] = hist['closedPnL_num'].abs()
hist['notional'] = hist['size_num'] * hist['price_num'].fillna(1)
```

```python
hist['date'] = pd.to_datetime(hist['date']).dt.normalize()
fg_grouped = fg_grouped.set_index('date').asfreq('D').ffill().reset_index()
merged = pd.merge(hist, fg_grouped, on='date', how='left')
merged['classification'].fillna('Unknown', inplace=True)
```

```python
daily = merged.groupby('date').agg(
    trades = ('closedPnL_num', 'count'),
    total_volume = ('notional', 'sum'),
    avg_leverage = ('leverage_num', 'mean'),
    avg_pnl = ('closedPnL_num', 'mean'),
    median_pnl = ('closedPnL_num', 'median'),
    win_rate=('is_win', lambda x: x.sum()/x.count() if x.count()>0 else np.nan)
).reset_index()
```

```python
daily = daily.merge(fg_grouped, on='date', how='left')
daily['classification'].fillna('unknown', inplace=True)
```

```python
print("\n Daily aggregated trader activity (head) ")
display(daily.head(8))
```

```python
# --- Better Sentiment Summary ---
sentiment_summary = (
    merged.groupby('classification', as_index=False)
    .agg(
        total_trades=('closedPnL_num', 'count'),
        total_pnl=('closedPnL_num', 'sum'),
        avg_daily_pnl=('closedPnL_num', 'mean'),
        median_pnl=('closedPnL_num', 'median'),
        win_rate=('is_win', 'mean'),
        avg_leverage=('leverage_num', 'mean'),
        avg_volume_usd=('notional', 'mean'),
        avg_trade_size=('size_num', 'mean'),
    )
    .sort_values('total_pnl', ascending=False)
)

# Convert win rate to %
sentiment_summary['win_rate'] = sentiment_summary['win_rate'] * 100

# Format numbers neatly
sentiment_summary = sentiment_summary.round({
    'total_pnl': 2,
    'avg_daily_pnl': 2,
    'median_pnl': 2,
    'win_rate': 1,
    'avg_leverage': 2,
    'avg_volume_usd': 2,
    'avg_trade_size': 2
})

print("ðŸ“Š Sentiment Performance Summary:")
display(
    sentiment_summary.style
    .background_gradient(subset=['total_pnl'], cmap='Greens')
    .background_gradient(subset=['win_rate'], cmap='Blues')
    .format({
        'total_pnl': '${:,.0f}',
        'avg_daily_pnl': '${:,.2f}',
        'median_pnl': '${:,.2f}',
        'avg_volume_usd': '${:,.2f}',
        'avg_trade_size': '{:,.2f}',
        'win_rate': '{:.1f}%',
        'avg_leverage': '{:.2f}'
    })
)
```
